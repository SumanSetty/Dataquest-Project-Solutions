# Analyzing Star Wars Survey Data

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding = "ISO-8859-1")

star_wars.head(10)
```

## Removing null columns in RespondentID column  and converting the column from float to int 

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars["RespondentID"] = (star_wars["RespondentID"]).astype(int)
star_wars.head()
```

```python
# converting Have you seen any of the 6 films in the Star Wars franchise? column from string to Boolena ##

str2bool = {"Yes": True, "No": False}
cols = ["Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]

for each in cols:
    
    star_wars[each] = star_wars[each].map(str2bool)

star_wars.head()
```

## Converting columns 3 to 8 (including) from string to boolean and renaimg the columns 

```python
import numpy as np

movie2bool = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie2bool)
    

star_wars = star_wars.rename(columns = {
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
})
    
star_wars.head()
```

## Converting columns 9 through 14 (including 14) to float and renaming the coulmns 

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns = {
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6",
})
star_wars.head()
```

## Plotting the movie with highest ranking ##

```python
star_wars[star_wars.columns[9:15]].mean()

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

labels = star_wars.columns[9:15].values
bar_positions = np.arange(6) + 0.5
tick_positions = range(1,7)

fig, ax = plt.subplots()

x = bar_positions
y = star_wars[star_wars.columns[9:15]].mean()
ax.bar(x,y)
ax.set_xticks(tick_positions)
ax.set_xticklabels(labels, rotation = 90)
plt.title("Mean rank for each movie")

plt.show()


```

Steps we did so far:
1. Read the csv file
2. Deleted rows with Null in "RespondentID" column
3. Converted the 2nd and 3rd column from string to boolean and renamed the column names of columns 4,5,6,7,8,9
4. Converted the 10th and 11th columns from strings to Booleans and renamed columns 10,11,12,13,14,15 and changed them to float type
5. Plotted the movies to their rankings

Plot analysis:
The first 3 movies (prequel trilogy) are the most ranked movies. People liked the newer movies than the older ones.

## Plotting the charaters liked and disliked the most

```python
star_wars[star_wars.columns[15:29]].head()
```

```python
import numpy as np
import pandas as pd
star_wars = star_wars.rename(columns = {
    
"Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.": "Like/Dislike:Han Solo",
"Unnamed: 16": "Like/Dislike:Luke Skywalker",
"Unnamed: 17": "Like/Dislike:Princess Leia Organa",
"Unnamed: 18": "Like/Dislike:Anakin Skywalker",
"Unnamed: 19": "Like/Dislike:Obi Wan Kenobi",
"Unnamed: 20": "Like/Dislike:Emperor Palpatine",
"Unnamed: 21": "Like/Dislike:Darth Vader",
"Unnamed: 22": "Like/Dislike:Lando Calrissian", 
"Unnamed: 23": "Like/Dislike:Boba Fett", 
"Unnamed: 24": "Like/Dislike:C-3P0", 
"Unnamed: 25": "Like/Dislike:R2 D2", 
"Unnamed: 26": "Like/Dislike:Jar Jar Binks", 
"Unnamed: 27": "Like/Dislike:Padme Amidala", 
"Unnamed: 28": "Like/Dislike:Yoda" 
})

star_wars[star_wars.columns[15:29]].head()
```

```python
characters_liked_counts = []
cols = ["Like/Dislike:Han Solo",
        "Like/Dislike:Luke Skywalker",
        "Like/Dislike:Princess Leia Organa",
        "Like/Dislike:Anakin Skywalker",
        "Like/Dislike:Obi Wan Kenobi",
        "Like/Dislike:Emperor Palpatine",
        "Like/Dislike:Lando Calrissian", 
        "Like/Dislike:Boba Fett", 
        "Like/Dislike:C-3P0", 
        "Like/Dislike:R2 D2", 
        "Like/Dislike:Jar Jar Binks", 
        "Like/Dislike:Padme Amidala", 
        "Like/Dislike:Yoda" ]
for each in cols:
    likes_dislikes = (star_wars[each] == "Very favorably").value_counts()
    likes = likes_dislikes[True]
    characters_liked_counts.append(likes)

print(characters_liked_counts)        
```

## Plotting characters liked the most


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

labels = ["Han Solo",
          "Luke Skywalker",
          "Princess Leia Organa",
          "Anakin Skywalker",
          "Obi Wan Kenobi",
          "Emperor Palpatine",
          "Lando Calrissian", 
          "Boba Fett", 
          "C-3P0", 
          "R2 D2", 
          "Jar Jar Binks", 
          "Padme Amidala", 
          "Yoda" ]
tick_positions = range(1,14)
bar_positions = np.arange(13) + 0.5

fig, ax = plt.subplots()
x = bar_positions
y = characters_liked_counts
plt.bar(x,y)
ax.set_xticks(tick_positions)
ax.set_xticklabels(labels, rotation = 90)
plt.title("Characters most liked")
plt.show()


```

## Analysis:
Han Solo and Yoda are the most liked characters, followed by Obi Wan Kenobi. Emporor Palpatine and Jar Jar Binks fall to the bottom of the liked list.
