# Visualizing The Gender Gap In College Degrees

## Plotting across STEM, Liberal Arts and Other Categories ##

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)


## Plotting STEM Categories ##

stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
fig = plt.figure(figsize=(16,20))

for sp in range(0,18,3):
    ax = fig.add_subplot(6,3,sp+1)
    sp_2 = int(sp/3)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp_2]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp_2]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp_2])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')
    
    if sp_2 == 5:
        ax.tick_params(labelbottom = 'on')
    ax.set_yticks([0,100])
    
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha =0.3)
    
    if sp_2 == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp_2 == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        

## plotting Liberal Arts Catergories ##

lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism',
                'Art and Performance', 'Social Sciences and History']

for sp in range(1,16,3):
    ax = fig.add_subplot(6,3,sp+1)
    sp_3 = int((sp-1)/3)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp_3]], c=cb_dark_blue, label='Women', linewidth=3 )
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[sp_3]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp_3])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')
    
    if sp_3 == 4:
        ax.tick_params(labelbottom = 'on')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp_3 == 0:
        ax.text(2005, 78, 'Women')
        ax.text(2002, 18, 'Men')
    elif sp_3 == 4:
        ax.text(1975, 65, 'Men')
        ax.text(1975, 30, 'Women')
        

## Plotting the Other categories ##
        
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

for sp in range(2,18,3):
    sp_4 = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp_4]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[sp_4]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp_4])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')
    if sp_4 == 5:
        ax.tick_params(labelbottom = 'on')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp_4 == 0:
        ax.text(2005, 78, 'Women')
        ax.text(2002, 18, 'Men')
    elif sp_4 == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2005, 30, 'Women')
plt.savefig("gender_degrees.png")    
plt.show()        
```
