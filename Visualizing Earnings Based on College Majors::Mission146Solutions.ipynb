# Visualizing Earnings Based On College Majors


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0], "\n")
print(recent_grads.head(), "\n\n")
print(recent_grads.tail(), "\n")
recent_grads.describe()

## Data Cleaning - drop rows with missing values ##

#Number of rows - Raw
raw_data_count = len(recent_grads.index)
print("Raw data row count:", raw_data_count, "\n")

#Drop NA
recent_grads = recent_grads.dropna()

#Number of rows - Cleaned data
cleaned_data_count = len(recent_grads.index)
print("Cleaned data row count:", cleaned_data_count, "\n")

```

## Data exploration using scatter plot ##

## Between Sample_size and Median ##

```python
recent_grads.plot(x = "Sample_size", y = "Median", kind = "scatter",
                 title = "Sample_size Vs Median")
```

# Between Sample_size and Median #

```python
recent_grads.plot(x = "Sample_size", y = "Unemployment_rate", kind = "scatter",
                 title = "Sample_size Vs Unemployment_rate")
```

# Between Full_time and Median #

```python
recent_grads.plot(x = "Full_time", y = "Median", kind = "scatter",
                 title = "Full_time Vs Median")
```

# Between ShareWomen and Unemplotment_rate #

```python
recent_grads.plot(x = "ShareWomen", y = "Unemployment_rate", kind = "scatter",
                 title = "ShareWomen Vs Unemployment_rate")
```

# Between Men and Median #

```python
recent_grads.plot(x = "Men", y = "Median", kind = "scatter",
                 title = "Men Vs Median")

```

# Between Women and Median #

```python
recent_grads.plot(x = "Women", y = "Median", kind = "scatter",
                 title = "Women Vs Median")
```

## Data Exploration using Histograms ##

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", 
        "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(8,15))
for r in range(0,4):
    ax = fig.add_subplot(4,1,r+1)
    ax = recent_grads[cols[r]].hist(bins= 25, range = (0,5000))
    plt.title(cols[r])
    
#Look further down for a seperate Median plot    
```
